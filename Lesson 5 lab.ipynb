```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['A'] = df.apply(lambda x : x['A'] + 10, axis = 1)
df
```

```python
# Calculate median salary
median_salary = df['Salary'].median()

# Function to categorize salary
def salary_category(salary):
    return 'Above Median' if salary > median_salary else 'Below Median'



# Applying functions
df['Salary Category'] = df['Salary'].apply(salary_category)


print(df)
```

```python
customers_data = {
    'customer_id': [1, 2, 3, 4],
    'customer_name': ['Alice', 'Bob', 'Charlie', 'David'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', 'david@email.com']
}

orders_data = {
    'order_id': [101, 102, 103, 104],
    'customer_id': [1, 2, 1, 1],
    'order_date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04']
}
```

```python
def find_customers_without_orders(customers_data, orders_data):
    # Find customer IDs with no orders
    customers_without_orders = [customer_id for customer_id in customers_data['customer_id'] if customer_id not in orders_data['customer_id']]

    # Filter customer data for customers without orders
    result = {key: [customers_data[key][i] for i in range(len(customers_data['customer_id'])) if customers_data['customer_id'][i] in customers_without_orders] for key in customers_data}

    return result

# Test the function
result = find_customers_without_orders(customers_data, orders_data)
print(result)
```
